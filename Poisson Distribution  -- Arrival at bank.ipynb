# Poisson Distribution
## Problem Statement 
Given that a bank has a footfall(arriving) of 6 people per minute, find out the

a) probability distribution for upto 20 people per minute<br> b) probability of at least 4 people per minute

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rate = 6
```

```python
n = np.arange(0,20)
n
```

```python
poisson  = stats.poisson.pmf(n,rate) # pmf -- probability mass function....  . 
```

```python
poisson
```

```python
plt.plot(n,poisson,'o-')
plt.title("Poisson $\lambda$ = %d" % rate)
plt.xlabel("Number of custmer arriving per minute")
plt.ylabel("Perobability of customer arriving per minute")
plt.show()
```

```python
#exacly 4 people 

poisson[4]
```

```python
# atleast 4 people each minute p(X>=4)

1-(poisson[0] + poisson[1] + poisson[2] + poisson[3] )

```
